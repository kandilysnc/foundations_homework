```python
import pandas as pd
%matplotlib inline
```

```python
df = pd.read_csv('1980.csv', encoding='latin-1')
df.head()
```

```python
df.columns = df.columns.str.replace("SE_","") #jangan lupa setiap ganti sesuatu replace safe it dengan variable baru. apa yang mau diganti ditaruh di bagian kiri. df.columns
df.head()
```

```python
codes = pd.read_csv('1980.txt', sep=':   ', skiprows=41, nrows=12, names=['code','name'])
codes
#kalau mau ganti nama column pakai names=['code', 'name']
```

```python
codes['name']= codes.name.str.strip()
#remove blank space from both sides of our name
codes_dict=dict(zip(codes.code,codes.name))
codes_dict
#buat jadi dictionary
```

```python
df=df.rename(columns=codes_dict)
df.head()
```

```python
df['pct_white'] = df['Total Population: White'] / df['Total Population']
df.head()
```

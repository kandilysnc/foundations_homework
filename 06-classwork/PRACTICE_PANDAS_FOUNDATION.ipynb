```python
import pandas as pd
import requests

%matplotlib inline
```

```python
response = requests.get('https://api.darksky.net/forecast/15074402c474e3dab67f7377e1f95519/37.8267,-122.4233')
data = response.json()
data
```

```python
data['hourly']['data']
```

```python
#if you have a list of dictionaries
#For loops
#list comprehensions
#for loops to count things or add things up

#now we can give pd.dataframe our list of dictionaries!
df = pd.DataFrame(data['hourly']['data'])
df.head()
```

```python
#find the hottest hour in the row

#make sure it is a list of dictionaries

df.sort_values(by='temperature', ascending=False).head(5) #top five the warmest temperature predictions over the next 48 hours
```

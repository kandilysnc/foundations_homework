```python
import requests
from bs4 import BeautifulSoup
```

```python
headers = {
    'User-Agent': 'Mozilla/5.0 (Macintosh; Intel Mac OS X 10_10_1) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/39.0.2171.95 Safari/537.36'
}

response = requests.get('https://www.goodreads.com/list/show/6934.Science_Fiction_Books_by_Female_Authors/', headers=headers)
doc = BeautifulSoup(response.text, 'html.parser')
```

```python
doc
```

```python
sorted_title = doc.find_all('td', attrs={'width':'100%'})
for title in sorted_title:
    titles=title.find('span', attrs={'itemprop':'name'})
    print(titles.text)
```

```python
sorted_author = doc.find_all('div', class_='authorName__container')
for author in sorted_author:
    names= author.find('span', attrs={'itemprop':'name'})
    print(names.text)
```

```python
sorted_ranks = doc.find_all('td', class_='number')
for rank in sorted_ranks:
    print(rank.text)
```

```python
sorted_rating = doc.find_all('span', class_='greyText smallText uitext')
for rating in sorted_rating:
    print(rating.text.strip())
```

```python
sorted_scores =doc.find_all('span', class_='smallText uitext')
for score in sorted_scores:
    score=score.find('a')
    print(score.text)

```

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
headers = {
    'User-Agent': 'Mozilla/5.0 (Macintosh; Intel Mac OS X 10_10_1) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/39.0.2171.95 Safari/537.36'
}

response = requests.get('https://www.goodreads.com/list/show/6934.Science_Fiction_Books_by_Female_Authors/', headers=headers)
doc = BeautifulSoup(response.text, 'html.parser')
```

```python
books_list = []
sorted1 = doc.find_all('tr')
#sorted = doc.find_all('span', class_='smallText uitext')
for item in sorted1:
    books = {}
    books['Rank'] = item.find('td', class_='number').text.strip()
    books['Title'] = item.find('a', class_='bookTitle').text.strip()
    books['Author'] = item.find('a', class_='authorName').text.strip()
    books['Rating'] = item.find('span', class_='greyText smallText uitext').text.strip()
    books['Vote'] = item.find('span', class_='smallText uitext').find_all('a')[1].text.strip()
    books['Score'] = item.find('span', class_='smallText uitext').find('a').text.strip()
    books_list.append(books)
    
print(books_list)
    
```

```python
df = pd.DataFrame(books_list)
df.head()
```

```python
df['Score'] = df['Score'].str.replace("score:", " ", case = False)
df.head()
```

```python
df['Vote'] = df['Vote'].str.replace("people voted", " ", case = False)
df.head()
```

```python
df['Rating'] = df['Rating'].str.replace("ratings", " ", case=False)
df.head()
```

```python
df["Rating"]= df["Rating"].str.replace("avg rating*.", " ", case = False)
df.head()
```

```python
df["Number of ratings"] = df['Rating'].str.extract(r"—(.*)", expand=True)
df.head()
```

```python
df['Rating'] = df['Rating'].str.replace(r"— (.*)", " ", case=False)
df.head()
                                        
```

```python
df['Series'] = df['Title'].str.extract(r"[(](.*),?#", expand=True)

df.head()
```

```python
df['Series'] = df['Series'].str.replace("," ," ")

df.head()
```

```python
df['Number in Series'] = df['Title'].str.extract(r"#(.*)[)]", expand=True)
df.head()
```

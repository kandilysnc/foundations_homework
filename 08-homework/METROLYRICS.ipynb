```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
headers = {
    'User-Agent': 'Mozilla/5.0 (Macintosh; Intel Mac OS X 10_10_1) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/39.0.2171.95 Safari/537.36'
}

response = requests.get('http://www.metrolyrics.com/bob-dylan-lyrics.html', headers=headers)
doc = BeautifulSoup(response.text, 'html.parser')
```

```python
bobdylan_list = []
sorted = doc.find_all('tr')

for item in sorted:
    bd = {}
    try:
        bd['Title'] = item.find('a', class_='title').text.strip()
        bd['Year'] = item.find_all('td')[2].text.strip() #use find_all karena pingin ke loop ke next step
        bd['URL'] = item.find(class_='title')['href']
        bd['Popularity'] = item.find(class_='bar').find('span')['style']
    except:
        pass
    
    bobdylan_list.append(bd)
    
print(bobdylan_list)










    
```

```python
df = pd.DataFrame(bobdylan_list)
df = df.dropna()
```

```python
df['Title'] = df['Title'].str.replace("Lyrics", " ", case = False)
df['Popularity'] = df['Popularity'].str.replace("width:", " ", case = False)
df['Popularity'] = df['Popularity'].str.replace("%", " ", case=False)
df['Popularity'] = df['Popularity'].str.replace(";", " ", case=False)
df.head()

```

```python
df.dtypes
```

```python
df['Popularity'] = df.Popularity.astype(float).round()
df['Popularity'] = df.Popularity.astype(int)
```

```python
def get_lirics(url):
    response = requests.get(url)
    doc = BeautifulSoup(response.content, 'html.parser')
    try:
        lyrics = doc.find_all('p', attrs={"class":'verse'})
        lirics = '\n'.join([line.text for line in lyrics])
        return lirics
    except:
        pass
```

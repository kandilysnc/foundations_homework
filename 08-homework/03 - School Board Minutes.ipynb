````python
minerals = []
pdfs = doc.find_all('p')[19:78]
for link in pdfs:
    if link.a is not None:
        meeting = {}
        meeting['link'] = 'http://www.mineral.k12.nv.us' + link.find('a')['href']
        meeting['date'] = link.get_text(strip=True)
        minerals.append(meeting)
        
print(minerals)    

#None special operator in python means does not exist
#get_text(strip=True) to clean any unwanted characters - to clean the data 
````
